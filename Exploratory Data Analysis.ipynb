**Cleaning the Data**

```python
anz_exp.isnull().sum()
```

```python
anz_exp_new=anz_exp.drop(["currency","country","merchant_code","bpay_biller_code","merchant_long_lat"],axis="columns")
```

```python
anz_exp_new
```

```python
anz_exp1=anz_exp_new.drop(["long_lat","txn_description","merchant_suburb"],axis="columns")
```

```python
anz_exp1
```

```python
anz_exp1.isnull().sum()
```

```python
anz_exp1=anz_exp1[~(anz_exp1.balance<20)]
```

```python
anz_exp1
```

**Understanding relationship between Features**

```python
correlation=anz_exp1.corr()
```

```python
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True)
```

```python
sns.pairplot(anz_exp1)
```

```python
plt.figure(figsize=(8,6))
plt.title('Transaction Status')
sns.countplot(anz_exp1['status'])
```
